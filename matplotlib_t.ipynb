## Bar plot 

```python
categories = ['A','B','C','D'] 
values = [10, 8, 16, 20]

plt.bar(categories, values, color = 'green')
plt.title('bar plot')
plt.xlabel('categories')
plt.ylabel('values')
```

## Subplot

```python
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axs = plt.subplots(2,1)

axs[0].plot(x,y1,'r')
axs[0].set_title('sin function')

axs[1].plot(x,y2, 'b')
axs[1].set_title('cos function')

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```
